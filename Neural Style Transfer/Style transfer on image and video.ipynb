```python
from os import listdir
from os.path import isfile, join
```

```python
import cv2
```

```python
# load neural transfer model
model_file_path = './models/'
model_file_paths = [f for f in listdir(model_file_path) if isfile(join(model_file_path, f))]
```

```python
# load image
img = cv2.imread('./images/Yuki.jpg')
```

```python
# load transfer style model
model = ('la_muse.t7')
```

### Work with image

```python
# apply each model style to our test image
for i in model_file_paths:
    style = cv2.imread('./art/' + str(model)[:-3] + '.jpg')
    neural_style_model = cv2.dnn.readNetFromTorch(model_file_path + model)  # load neural transfer style model
    
    height, width = int(img.shape[0]), int(img.shape[1])  
    new_width = int((640/height) * width)
    resized_img = cv2.resize(img, (new_width, 640), interpolation=cv2.INTER_AREA)  # resize to fix height
    
    R_mean, G_mean, B_mean = 103.93, 116.77, 123.68 # mean values for ImageNet training set
    
#     print(R_mean)
    
    inp_blob = cv2.dnn.blobFromImage(resized_img, 1.0, (new_width, 640),
                                    (R_mean, G_mean, B_mean), swapRB = False, crop = False)
    
    neural_style_model.setInput(inp_blob)
    output = neural_style_model.forward()
    
    output = output.reshape(3, output.shape[2], output.shape[3])
    
    # re-order the channels
    output[0] += R_mean
    output[1] += G_mean
    output[2] += B_mean
    output /= 255
    output = output.transpose(1,2,0)
    
    # print original, style and result images
    
    cv2.imshow('Original', img)
    cv2.imshow('Style', style)
    cv2.imshow('Neural Transfer Style', output)
    cv2.waitKey(0)
    
    # Close cycle using key
    
    if cv2.waitKey(0) & 0xFF == 27:
        break
        
# Destroy windows
cv2.destroyAllWindows()    
    
```

### Let's proceed to the video

```python
rgb_mean=(R_mean, G_mean, B_mean)
```

```python
def neural_style_transfer(img, size=320, model='mosaic.t7', upscale=1, rgb_mean=(R_mean, G_mean, B_mean)):
    """
    Function for image neural style transfer
    """
    
    model_file_path = 'models/'
    style = cv2.imread('art/' + str(model)[:3] + '.jpg')
    neural_style_model = cv2.dnn.readNetFromTorch(model_file_path + model)  # load neural transfer style model
    
    height, width = int(img.shape[0]), int(img.shape[1])  
    new_width = int((size/height) * width)
    resized_img = cv2.resize(img, (new_width, size), interpolation=cv2.INTER_AREA)  # resize to fix height
    
    inp_blob = cv2.dnn.blobFromImage(resized_img, 1.0, (new_width, size),
                                    rgb_mean, swapRB = False, crop = False)
    
    neural_style_model.setInput(inp_blob)
    output = neural_style_model.forward()
    
    output = output.reshape(3, output.shape[2], output.shape[3])
    
    # re-order the channels
    output[0] += rgb_mean[0]
    output[1] += rgb_mean[1]
    output[2] += rgb_mean[2]
    output /= 255
    output = output.transpose(1,2,0)
    
    output = cv2.resize(output, None, fx=upscale, fy=upscale, interpolation=cv2.INTER_LINEAR)
    
    return output
    
```

```python
# Video opening

# cap = cv2.VideoCapture('./video/run.mp4')
cap = cv2.VideoCapture(0)  # TO RUN YOUR WEBCAM

while True:    
    ret, frame = cap.read()  # frame read
    cv2.imshow('Video style transfer', neural_style_transfer(frame,
                                                             size=320,
                                                             model='starry_night.t7',
                                                             upscale=1))
    # frame closing
    if cv2.waitKey(1) & 0xFF == 27:  # press 'esc' button to close
        break
        
# Destroy windows
cap.release()
cv2.destroyAllWindows()
```

### Working with green screen

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

%matplotlib inline
```

```python
# read the image
image = cv2.imread('./images/car_green_screen.jpg')

# look at the image type and dimensions

print('Image type:', type(image))
print('Image dimensions', image.shape)
```

```python
img_copy = np.copy(image) # copy of the image
img_copy = cv2.cvtColor(img_copy,cv2.COLOR_BGR2RGB) # switch color

plt.imshow(img_copy) # print the copied image with changed color
# plt.imshow(image)
```

```python
# green background isolation

lower_green = np.array([0, 200, 0])
upper_green = np.array([250, 255, 250])
```

```python
# mask of the area
mask = cv2.inRange(img_copy, lower_green, upper_green)

plt.imshow(mask, cmap='gray'); # print the mask
```

```python
masked_image = np.copy(img_copy)

masked_image[mask != 0] = [0,0,0] # if there is no mask, there black pixels
plt.imshow(masked_image); # print the mask
```

```python
background = cv2.imread('./images/london.jpg') # background image loading
background = cv2.cvtColor(background, cv2.COLOR_BGR2RGB) # change color
background = background[0:masked_image.shape[0],0:masked_image.shape[1]] # change shape
background[mask == 0] = [0,0,0] # mask the background

plt.imshow(background)
```

Image moving

```python
# x,y,h,w
x = 390
y = 300
h = masked_image.shape[0]
w = masked_image.shape[1]

background = cv2.imread('./images/london.jpg')
background = cv2.cvtColor(background, cv2.COLOR_BGR2RGB) # change color
background = background[y:y+h, x:x+w] # change shape
background[mask == 0] = [0,0,0] # mask the background

plt.imshow(background); # print image
```

```python
# combine two images:
super_image = masked_image + background
plt.imshow(super_image);
```
